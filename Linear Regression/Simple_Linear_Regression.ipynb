```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,5,10,8,17])
y = np.array([34,30,27,20,25,10])
```

Now, we will see the importance of choosing learning rate effectively.

```python
A,B = 1,1
lr = 0.015
# little bit higher learning rate than the optimal rate
```

```python
for i in range(500):
  y_pred = x*A + B
  dA = (2*(y_pred - y)*x).sum()/6
  dB = (2*(y_pred - y)).sum()/6
  A = A - dA*lr
  B = B - dB*lr
  error = ((y_pred - y)**2).sum()/6
  print("After {}th iteration, the error is".format(i+1),error)
```
